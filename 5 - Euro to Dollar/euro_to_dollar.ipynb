```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
%matplotlib inline

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

```python
exchange_rates.rename(columns={'[US dollar ]':'US_dollar',
                      'Period\\Unit:':'Time'}, inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates = exchange_rates.reset_index(drop=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates.loc[:, ['Time', 'US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
```

# Idea for Storytelling
We'll show up the variation of the euro-dollar rate under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017) and Donald Trump (2017-2021)) and under significant events over each mandate.

Thus, we'll remove the data before 01-01-2001 and the few days in the year 2021:

```python
euro_to_dollar['Year'] = euro_to_dollar['Time'].dt.year
euro_to_dollar = euro_to_dollar[euro_to_dollar['Year'].between(2001, 2020)]
euro_to_dollar = euro_to_dollar.reset_index(drop=True)
euro_to_dollar
```

```python
style.use('fivethirtyeight')

# Separating the three periods
Bush = euro_to_dollar[euro_to_dollar['Year'].between(2001, 2008)]
Obama = euro_to_dollar[euro_to_dollar['Year'].between(2009, 2016)]
Trump = euro_to_dollar[euro_to_dollar['Year'].between(2017, 2020)]

# Ploting
fig, ax = plt.subplots(figsize=(17,12))
ax.plot(Bush['Time'], Bush['rolling_mean'],
        linewidth=2.5, color='#c519be')
ax.plot(Obama['Time'], Obama['rolling_mean'],
        linewidth=2.5, color='#df3c1f')
ax.plot(Trump['Time'], Trump['rolling_mean'],
        linewidth=2.5, color='#2f56e3')
ax.grid(alpha=0.4)

# Title and subtitle
ax.text(730360, 1.65, s='EURO-USD rate averaged 1.22 and increases 37% under the last three US presidents',
        weight='bold', fontsize=23)
ax.text(730360, 1.62, s='EURO-USD exchange rates jumps up 50% (Bush), decreases 20% (Obama) and remains stable with Trump (+15%)',
        fontsize=19)

# Important Historic Events
ax.text(731000, 0.84, s='Afghanistan War', fontsize=19)
ax.axhline(0.85, xmin=0.025, xmax=0.06, color='black', linewidth=1)

ax.text(731650, 1.07, s='Iraq War', fontsize=19)
ax.axhline(1.08, xmin=0.114, xmax=0.149, color='black', linewidth=1)

ax.text(732555, 1.215, s='Financial Crisis', fontsize=19)
ax.axvline(732650, ymin=0.555, ymax=0.590, color='black', linewidth=1)

ax.text(733400, 1.53, s='Greek Crisis', fontsize=19)
ax.axvline(733750, ymin=0.865, ymax=0.900, color='black', linewidth=1)

ax.text(733500, 1.14, s='Obamacare', fontsize=19)
ax.axvline(733890, ymin=0.47, ymax=0.66, color='black', linewidth=1)

ax.text(734000, 1.47, s='End of Iraq War', fontsize=19)
ax.axvline(734460, ymin=0.70, ymax=0.82, color='black', linewidth=1)

ax.text(735580, 1.3, s='End of Afghanistan War', fontsize=19)
ax.axvline(735580, ymin=0.565, ymax=0.605, color='black', linewidth=1)

ax.text(736900, 1.03, s='Coronavirus', fontsize=19)
ax.axvline(737570, ymin=0.330, ymax=0.36, color='black', linewidth=1)

ax.legend(['Bush', 'Obama', 'Trump'], fontsize=20)
plt.show()
```
